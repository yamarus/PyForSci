 ``try .. except``   . 

```python
k = 1/0 
```

```python
def division(a,b):
    try: 
        r = a/b
    except ZeroDivisionError:
        r = float('inf')
    except Exception as ex:
        print(ex.args)
        r = float('nan')
    else:
        print(' ')
    finally:
        return r
    

print(division(1,0))
```

 ``try .. except``         .

```python
fname='example1_file.txt'
f=open(fname, 'r')
print(f.read())
f.close()
```

```python
fname='example1_file.txt'
try:
    f=open(fname, 'r')
except:
    print(f'    {fname}')
else: 
    print(f.read())
finally: f.close()
```
